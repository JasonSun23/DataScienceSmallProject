```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as py
%matplotlib inline
#-------------------------------
# Your code goes here.
# Add as many cells as you need
#-------------------------------

#Q1 

#Read the following three files as pandas dataframe
unames = ['user_id', 'gender', 'age', 'occupation', 'zipc']
users = pd.read_table('ml-1m/users.dat', sep='::', header=None, names=unames, engine='python')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('ml-1m/ratings.dat', sep='::', header=None, names=rnames, engine='python')

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('ml-1m/movies.dat', sep='::', header=None, names=mnames, engine='python')

#Merge all of the data into a single Pandas DataFrame
data = pd.merge(pd.merge(ratings, users), movies)

#Store the data into an HDF5 file.
store = pd.HDFStore('data.h5')
store['data'] = data
data
```

```python
#Q2

#Compute how many movies have an average rating over 4.5 overall.
mean_ratings = data.pivot_table('rating', index='title', aggfunc='mean')
active_titles = mean_ratings.index[mean_ratings > 4.5]
mean_rating_fpf = mean_ratings.ix[active_titles]
print('Rating over 4.5:', len(mean_rating_fpf))
```

```python
#Compute how many movies have an average rating over 4.5 overall among men and women.
mean_ratings_gender = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
active_titlesM = mean_ratings_gender.index[mean_ratings_gender['M'] > 4.5]
active_titlesF = mean_ratings_gender.index[mean_ratings_gender['F'] > 4.5]

mean_ratings_genderM_fpf = mean_ratings_gender.ix[active_titlesM]
mean_ratings_genderF_fpf = mean_ratings_gender.ix[active_titlesF]
print('Rating over 4.5 among Male:', len(mean_ratings_genderM_fpf))
print('Rating over 4.5 among Female:', len(mean_ratings_genderF_fpf))
```

```python
#Compute how many movies have an median rating over 4.5 overall among men and women over age 30.
data_over_age = data[data.age > 30]
median_ratings_gender_age = data_over_age.pivot_table('rating', index='title', columns='gender', aggfunc='median')
active_titlesM = median_ratings_gender_age.index[median_ratings_gender_age['M'] > 4.5]
active_titlesF = median_ratings_gender_age.index[median_ratings_gender_age['F'] > 4.5]
median_ratings_genderM_age_fpf = median_ratings_gender_age.ix[active_titlesM]
median_ratings_genderF_age_fpf = median_ratings_gender_age.ix[active_titlesF]
print('Rating over 4.5 among Male over age 30:', len(median_ratings_genderM_age_fpf))
print('Rating over 4.5 among Female over age 30:', len(median_ratings_genderF_age_fpf))
```

```python
#We define population is the movie with most number of ratings. Then, we find it out.
popularity = data.groupby('title').size()
print(popularity.sort_values(ascending=False)[0:10])
```

```python
#1. We conjecture that people whose occupation is "10:K-12 student" are the easiest to please.

#Compute the ratings mean and standard deviation of grouped by occupations
#Our result show that occupation 10 "k-12 student" have the highest average movie rating among different movies. So it
#prove our conjecture.
occu= data.pivot_table('rating', index='title', columns='occupation', aggfunc='mean')
occu = occu.dropna()
result = pd.concat([occu.mean(axis=0), occu.std(axis=0)], axis=1)
result.columns = ['Mean', 'Std']
print(result.sort_values(by='Mean', ascending=False).head(5))
```

```python
#2. We conjecture that Female are easier to please.

#Compute the ratings mean and standard deviation of grouped by gender
#Our result show that Female have higher average movie rating among different moview. So it prove our conjecture.
#But, the female have bigger standard deviation, which means their rating have more variance.
gen = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
gen = gen.dropna()
result = pd.concat([gen.mean(axis=0), gen.std(axis=0)], axis=1)
result.columns = ['Mean', 'Std']
print(result.sort_values(by='Mean', ascending=False))
```

```python
#3. We conjecture users whose age belong to "1:1-12" are the easiest to please.

#Compute the ratings mean and standard deviation of grouped by age
#Our result show that people whose age between 50 and 56 have the largest average movie rating among different movies.
#So, it seems that people aged between 50 and 56 are the easiest group of people to please.
#And people whose age between 1 and 12 is at the second place. But, they have quite large variance, which means their rating have more variance.
age = data.pivot_table('rating', index='title', columns='age', aggfunc='mean')
age = age.dropna()
result = pd.concat([age.mean(axis=0), age.std(axis=0)], axis=1)
result.columns = ['Mean', 'Std']
print(result.sort_values(by='Mean', ascending=False))
```

```python
#1. We conjecture that the occupatin "8:farmer" have the largest variance (more extreme ratings) of the average rating for movies.

#Plot a histogram of the average rating for movies among different occupations
#Compared with other occupations, Occupation 8 "farmer" looks like a bi-modal distribution. And, obviously, it has the largest
#variance with more extreme ratings, which prove our conjecture.
for i in range(21):
    mean = occu[i].mean(axis=0)
    std = occu[i].std(axis=0)
    py.subplot(5, 5, i+1)
    plt.hist(occu[i])
    plt.xlabel('Rating', fontsize=8)
    plt.ylabel('Frequency', fontsize=8)
    plt.title(r'Occupation: %d $\mu=%2f,\ \sigma=%2f$' %(i, mean, std), fontsize=10)
plt.subplots_adjust(bottom=0.1, right=3, top=5.0, left=0.1)
plt.show()
```

```python
#2. We conjecture that the women have more extreme ratings.

#Plot a histogram of the average rating for movies between men and women.
#Obvioulsy, female have more extreme ratings among different movies, and its standard devieation is bigger than male.
#So, it prove our conjecture.
j=1
for i in ['M', 'F']:
    mean = gen[i].mean(axis=0)
    std = gen[i].std(axis=0)
    py.subplot(1, 2, j)
    plt.hist(gen[i], color='green')
    plt.xlabel('Rating', fontsize=8)
    plt.ylabel('Frequency', fontsize=8)
    plt.ylim(0,900)
    plt.title(r'Gender: %s $\mu=%2f,\ \sigma=%2f$' %(i, mean, std), fontsize=11)
    j = j+1
plt.subplots_adjust(bottom=0.1, right=1.2, top=0.9, left=0.1)
plt.show()
```

```python
#3. We conjecture that the age range "1:Under 18", "56:56+" have more extreme ratings.

#Plot a histogram of the average rating for movies among different ages
#Obvioulsy, the people age between 1 and 18  have more extreme ratings among different movies, and its standard devieation is bigger than other age groups.
#So, it prove our conjecture.
j=1
for i in [1, 18, 25, 35, 45, 50, 56]:
    mean = age[i].mean(axis=0)
    std = age[i].std(axis=0)
    py.subplot(3, 3, j)
    plt.hist(age[i])
    plt.xlabel('Rating', fontsize=8)
    plt.ylabel('Frequency', fontsize=8)
    plt.title(r'Age: %s $\mu=%2f,\ \sigma=%2f$' %(i, mean, std), fontsize=10)
    j = j+1
plt.subplots_adjust(bottom=0.1, right=1.8, top=3.0, left=0.1)
plt.show()
```
